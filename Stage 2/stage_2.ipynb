```python
census_nsw = '2016_census_nsw.csv'
census_vic = '2016_census_vic.csv'
```

```python
def get_age_proportion(filepath):
    census = pd.read_csv(filepath)
    proportion_20_44 = {}
    p_20_24 = census['Age_20_24_yr_P']
    p_25_24 = census['Age_25_34_yr_P']
    p_35_44 = census['Age_35_44_yr_P']
    t_p = census['Tot_P_P']
    sa2 = census['SA2_MAINCODE_2016']

    for i in range(len(sa2)):
            
        sum_p = p_20_24[i]+p_25_24[i]+p_35_44[i]
        
        print(i,p_20_24[i],p_25_24[i],p_35_44[i],sum_p,t_p[i],sum_p/t_p[i])
        
        
        proportion_20_44[sa2[i]] = sum_p/t_p[i]
    return proportion_20_44
```

```python
proportion_20_44_nsw = get_age_proportion(census_nsw)
proportion_20_44_vic = get_age_proportion(census_vic)
```

```python
conn.execute("DROP TABLE IF EXISTS vulnerabilitydata_raw")
# pgquery (conn, "DROP TABLE IF EXISTS measuredata", None, "Reset Table measuredata")
vulnerabilitydata_raw = """
CREATE TABLE vulnerabilitydata_raw (
    area_id SERIAL PRIMARY KEY,
    area_name VARCHAR(200),
    population_density FLOAT,
    healthservices_density FLOAT,
    confirmedcase INT
)
"""
conn.execute(vulnerabilitydata_raw)

conn.execute("""
    INSERT INTO vulnerabilitydata_raw
        SELECT  R.area_id,
            R.area_name,
            ((PA.population_density-M.avg_population_density)/stddev_population_density),
            ((HA.count_healthservices-M.avg_healthservices_density)/stddev_healthservices_density), 
            confirmedcase
        FROM regions AS R 
            LEFT JOIN population_by_area AS PA ON R.area_id = PA.area_id
            LEFT JOIN healthservices_area AS HA ON R.area_id = HA.area_id
            LEFT JOIN case_postcode AS CP ON R.area_id = CP.area_id
            , measuredata M
""")


conn.execute("""UPDATE vulnerabilitydata_raw SET population_density = 0 WHERE population_density IS NULL;""")
conn.execute("""UPDATE vulnerabilitydata_raw SET healthservices_density = 0 WHERE healthservices_density IS NULL;""")
conn.execute("""UPDATE vulnerabilitydata_raw SET confirmedcase = 0 WHERE confirmedcase IS NULL;""")

pd.read_sql_query('SELECT * FROM vulnerabilitydata_raw', conn)
```
